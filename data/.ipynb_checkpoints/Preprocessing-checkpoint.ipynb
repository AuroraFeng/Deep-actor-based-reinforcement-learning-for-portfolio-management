````python
import numpy as np
import pandas as pd
from IPython.display import display # Allows the use of display() for DataFrames
#from mymath import *
from pylab import *
import matplotlib.pyplot as plt

# Import supplementary visualizations code visuals.py
#import visuals as vs

# Pretty display for notebooks
%matplotlib inline

from datetime import time
````

````python
# load data to dataframe:
def get_df(filename):
    df = pd.read_csv(filename,error_bad_lines=False, warn_bad_lines=True)
    return df
````

````python
df_SP500 = get_df('S&P500.csv')#pd.read_csv('S&P500.csv',error_bad_lines=False, warn_bad_lines=True)
df_Target = get_df('Target.csv')
df_VIX = get_df('VIX.csv')
df_Walmart = get_df('Walmart.csv')
display(df_SP500.head())
#display(df_Target.head())
#display(df_VIX.head())
#display(df_Walmart.head())
````

````python
def scattered_plot(df, features, nrow=1, figsize=(20,3)):
    # df is a dataframe; features is a list of column names
    nf = len(features)
    ncol = (nf+nrow-1)/nrow
    fig, axes=plt.subplots(nrow, ncol, figsize=figsize,squeeze=False)
    ind = 0
    for i in range(nrow):
        for j in range(ncol):
            if ind<nf:
                name=features[ind]
                axes[i][j].plot(df[name])
                axes[i][j].set_title(name)
                ind = ind+1
    plt.show()
````

````python
def holdon_plot(df, features, figsize=(8,3)):
    nf = len(features)
    plt.figure(figsize=figsize)
    #plt.hold(True)
    for i in range(nf):
        name=features[i]
        plt.plot(df[name])
    plt.legend(features)
    plt.show()
````

## We show the plot of ```Open```, ```Close```, ```Low``` and ```High```:

````python
features = ['Open', 'Close', 'Low', 'High']
````

````python
print "S&P500:"
scattered_plot(df_SP500, features, nrow=1)
#holdon_plot(df_VIX, features)
````

````python
print "VIX:"
scattered_plot(df_VIX, features, nrow=1)
#holdon_plot(df_VIX, features)
````

````python
print "Target:"
scattered_plot(df_Target, features, nrow=1)
#holdon_plot(df_VIX, features)
````

````python
print "Walmart:"
scattered_plot(df_Walmart, features, nrow=1)
#holdon_plot(df_VIX, features)
````
